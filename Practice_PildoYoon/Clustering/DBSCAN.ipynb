# 밀도 기반 클러스터링 기법인 DBSCAN을 실습하자

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
# 비구형 데이터를 생성하고 K-means 알고리즘을 수행하는 코드
from sklearn.datasets import make_moons
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# 1. 비구형 데이터 생성 (두 개의 반달 모양)
X_moons, y_moons = make_moons(n_samples=300, noise=0.1, random_state=42)

# 3. 비구형 데이터에 대한 K-means 결과 시각화
plt.figure(figsize=(8, 6))
plt.scatter(X_moons[:, 0], X_moons[:, 1])
plt.title('Moon data')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import DBSCAN
import numpy as np
import pandas as pd

# DBSCAN 모델 설정
db = DBSCAN(eps=0.3, min_samples=20).fit(X_moons)

# 결과 시각화
plt.figure(figsize=(8, 6))

# 군집 라벨에 따른 색상 설정
unique_labels = set(db.labels_)
colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_labels))]

for k, col in zip(unique_labels, colors):
    if k == -1:
        # 노이즈 포인트는 검은색으로 표시
        col = [0, 0, 0, 1]

    class_member_mask = (db.labels_ == k)
    xy = X_moons[class_member_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=tuple(col),
             markeredgecolor='k', markersize=10, label=f'Cluster {k}' if k != -1 else 'Noise')

plt.title('DBSCAN Clustering')
plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')
plt.legend()
plt.show()
```
